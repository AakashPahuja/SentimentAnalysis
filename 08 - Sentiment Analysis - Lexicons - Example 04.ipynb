```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
```

```python
#Read the datasets
path = './data/'
filePrefix = 'training_'

def apply_stopwording(corpus, min_len):
    punctuations = ".,\"-\\/#!?$%\^&\*;:{}=\-_'~()"    
    filtered_corpus = [token for token in corpus if (not token in stopwords.words('english') and len(token)>min_len) and (not token in punctuations)]
    return filtered_corpus

testing = []

fileName=path+"testing_sentiment.txt"
f=open(fileName,'r')
lines=f.readlines()
for line in lines:
    linetext = line.replace('\n',' ').lower()
    text,label = linetext.split("\t")
    tokens = nltk.word_tokenize(text)
    cleaned_tokens = apply_stopwording(tokens,3)
    testing.append((cleaned_tokens,label.rstrip().lower()))
f.close
```


## Reading the Lexicon Vocabulary

```python
def loadSentiWordNet(lfile):
    lf = open(lfile)
    lines=lf.readlines()
    lf.close
    lexicon = {}
    for line in lines:
        info = line.split("\t")
        try:
            p_score = float(info[2])
            n_score = float(info[3]) * -1.0
            words = info[4].split(" ")
            for word in words:
                term = word.split("#")
                lexicon[term[0]]= p_score + n_score
        except:
            pass
    return lexicon

lexicon_dictionary = "SentiWordNet_3.0.0_20130122.txt"
lex_fileName=path+lexicon_dictionary
lexicon_dictionary = loadSentiWordNet(lex_fileName)
```

```python
def scoreSentiment(text,lex_dic):
    score = 0.0
    sentiment = "Neutral"
    keys = lex_dic.keys()
    for word in text:
        if word in keys:
            score = score + lexicon_dictionary[word]
    
    if score>1.5:
        sentiment = "Positive"
    elif score<-1.5:
        sentiment = "Negative"
    return sentiment, score
```

```python
for test in testing:
    text = test[0]
    sent = test[1]
    p_sent, s_score = scoreSentiment(text,lexicon_dictionary)
    print ("%s [%s][%s]" % (p_sent,s_score,sent))
```
